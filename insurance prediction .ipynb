```python
# get dummy values

gender_new = pd.get_dummies(df['Gender'])
vehicle_age_new = pd.get_dummies(df['Vehicle_Age'])
vehicle_damage_new = pd.get_dummies(df['Vehicle_Damage'])


gender_test = pd.get_dummies(df0['Gender'])
vehicle_age_test = pd.get_dummies(df0['Vehicle_Age'])
vehicle_damage_test = pd.get_dummies(df0['Vehicle_Damage'])
```

```python
# add dummy columns with the train and test sets
df = pd.concat([df,gender_new, vehicle_age_new, vehicle_damage_new], axis=1 )

df0 = pd.concat([df0,gender_test, vehicle_age_test, vehicle_damage_test], axis=1 )
```

```python
# drop all the old columns as dummy variables would be used at their places
df = df.drop( ['Gender' , 'Vehicle_Age' , 'Vehicle_Damage'], axis=1)

df0 = df0.drop( ['Gender' , 'Vehicle_Age' , 'Vehicle_Damage'], axis=1)
```

```python
df.head()
```

```python
# categories x and y

x = df.drop(['Response'], axis=1)
y = df['Response']
x_test = df0
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x = scalar.fit_transform(x)

x_test = scalar.fit_transform(x_test)
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
model1 = LinearRegression().fit(x_train,y_train)
```

```python
model1.predict(x_val)
```

```python
# XGBClassifier
from xgboost import XGBClassifier
model2 = XGBClassifier().fit(x_train, y_train)
model2.predict(x_val)
pred2 = model2.predict(x_test)
```

```python
# DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
model3 = DecisionTreeRegressor().fit(x_train , y_train)
pred3 = model3.predict(x_test)
```

```python
#LogisticRegression
from sklearn.linear_model import LogisticRegression
model4 = LogisticRegression().fit(x_train , y_train)
pred4 = model4.predict(x_test)
pred4
```

```python
#DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier 
model5 = DecisionTreeClassifier(random_state = 100,max_depth=3, min_samples_leaf=5).fit(x_train , y_train)
pred5 = model5.predict(x_test)
pred5
```

```python
#GaussianNB this algorithm gave the highest accuracy with 78% on AV hackathon.
from sklearn.naive_bayes import GaussianNB
model6 = GaussianNB().fit(x_train , y_train)
pred6 = model6.predict(x_test)
pred6
```

```python
#CatBoostClassifier
from catboost import CatBoostClassifier
model8 = CatBoostClassifier().fit(x_train , y_train)
pred7 = model7.predict(x_test)
pred7
```

```python
# Create columns for the new csv file 
output_df = pd.DataFrame({'id' : df0['id'].values})
output_df['Response'] = pred6 # due to highest accuracy we use this algorithm
```

```python
# submit file to csv 
output_df.to_csv('Submission.csv' , index=False)
```
